# Student Performance Analysis
This notebook loads, cleans, analyzes, and visualizes the student performance dataset (`student-mat.csv`).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"C:\Users\gajen\Downloads\student\student-mat.csv",sep=';')
print(df.head())
df.head()
```

## Explore and Clean Data

```python
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
print("Duplicate rows:",df.duplicated().sum())
df=df.drop_duplicates()
df.info()
```

## Analysis

```python
print("Average Final Grade:", round(df['G3'].mean(),2))
print("Students scoring above 15:", (df['G3']>15).sum())
print("Correlation studytime vs G3:", df['studytime'].corr(df['G3']))
gender_avg=df.groupby('sex')['G3'].mean()
print(gender_avg)
```

## Visualizations

```python
plt.figure(figsize=(6,4))
plt.hist(df['G3'],bins=10,edgecolor='black')
plt.title('Distribution of Final Grades')
plt.xlabel('G3'); plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df['studytime'],df['G3'])
plt.title('Study Time vs Final Grade')
plt.xlabel('Study Time'); plt.ylabel('G3')
plt.show()

plt.figure(figsize=(6,4))
gender_avg.plot(kind='bar')
plt.title('Average Final Grade by Gender')
plt.xlabel('Gender'); plt.ylabel('Average G3')
plt.show()
```

## Conclusion

- Dataset loaded and cleaned.
- Average final grade calculated.
- Counted students scoring above 15.
- Examined correlation between study time and performance.
- Compared average performance by gender using a bar chart.
